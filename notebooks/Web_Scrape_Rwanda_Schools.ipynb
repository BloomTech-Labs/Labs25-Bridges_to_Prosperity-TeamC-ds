#### Website to scrape
http://www.edorica.com/schools

```
pages = 624

school_dict = {'School Name': []}
location_dict = {'Province': [],
                 'District': [],
                 'Sector': []}



# go thru each page on the webpage
for i in range(1, 625):
  print(i)
  URL = f'http://www.edorica.com/schools?currentpage={i}'
  page = requests.get(URL)
  soup = BeautifulSoup(page.content, 'html.parser')

  # loop thru the em tags and append the value to the school dictionary.
  for i in soup.find_all('em'):
    school_dict['School Name'].append(i.get_text())

  # loop thru the li tags and append the locations to their respective 
  # keys in the location dictionary
  for i in soup.find_all('li'):
    if 'Location' in i.get_text():
      tmp = i.get_text().replace('Location: ', "")
      tmp = tmp.split(',')
      if len(tmp) == 3:
          
        location_dict['Province'].append(tmp[0])
        location_dict['District'].append(tmp[1])
        location_dict['Sector'].append(tmp[2])
      elif len(tmp) == 2:
        location_dict['Province'].append(tmp[0])
        location_dict['District'].append(tmp[1])
        location_dict['Sector'].append('Unknown')
      elif len(tmp) == 1:
        location_dict['Province'].append(tmp[0])
        location_dict['District'].append('Unknown')
        location_dict['Sector'].append('Unknown')
      elif len(tmp) == 0:
        pass

        

      # print(tmp[0])
      # print(tmp[1])
      # print(tmp[2])
    else:
      pass
  
```

```
loc_df = pd.DataFrame(location_dict)
school_df = pd.DataFrame(school_dict)

df = pd.concat([school_df, loc_df], axis=1)
print(df.shape)
df.head()
```
