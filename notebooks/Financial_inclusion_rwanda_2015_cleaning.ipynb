```
# install the tabula package and import needed libraries
!pip install tabula-py
import tabula
import pandas as pd
```

```
file_path = "/content/BKLT_FinancialInclusionRwanda_2016.pdf"
#Convert your file
df = tabula.read_pdf(file_path, pages=24)
```

```
# get the pandas dataframe from the Tabula object
df = df[0]

# # Make the column names equal to the mapper .
mapper = {'Unnamed: 0': 'District', 'Unnamed: 1': 'Destination', 'Umurenge': 'Umurenge SACCO', 'Unnamed: 2': 'MFIs', 'Unnamed: 3': 'Bank Branch',
       'Unnamed: 4': 'ATM', 'Unnamed: 5': 'Bank Agent', 'Unnamed: 6': 'Mobile Money Agent' }

df = df.rename(mapper=mapper, axis=1)

# drop the first two rows, since there is no data contained in those rows.
df = df.drop([0, 1])

# drop the destination column because it is all Nan Values
df = df.drop(columns='Destination')

df
```

```
# save to a csv
df.to_csv('avg_travel_time_financial_access.csv', index=False)
```
