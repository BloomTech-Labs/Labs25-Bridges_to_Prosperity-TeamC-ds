```python
import pandas as pd
import seaborn as sns
```

```python
# Read in Bridge to prosperity data, and save to a dataframe
file_path = '../B2P Rwanda Site Assessment Data_2020.06.03.csv'
df = pd.read_csv(file_path, encoding='latin-1')
print(df.shape)
df.head()
```

```python
# extract the dates from the 'Form: Form Name' column to its own date column

df['Form Date'] = df['Form: Form Name'].str.strip('Project Assessment - ')

# convert the New date column to Datetime format
df['Form Date'] = pd.to_datetime(df['Form Date'], infer_datetime_format=True)


df.head()
```

```python
df[df['Bridge Site Name'] == 'Gatare']
```

```python
# are there columns that are all null values?

for column in df.columns:
    print(f'{column}', df[column].isnull().sum())
```

```python
# drop columns that are all null values.
col_drop = ['Community Served 6', 'Community Served 7',
            'Community Served 8', 'Community Served 9',
            'Community Served 10', 'Country'
            ]

df = df.drop(columns=col_drop)
df.shape
```

```python
# melt the dataframe so all communities are in a single column. 
id_variables = ['Country', 'Province', 'District', 
                'Sector', 'Cell', 'Bridge Site Name',
                'Project Stage', 'Project Sub-Stage', 
                'Project Code', 'Bridge Type', ' Span (m)',
                ' GPS (Latitude)', 'GPS (Longitude)',
                'Individuals Directly Served', 'Form: Form Name',
                'CaseSafeID Form', 'Bridge Opportunity: Opportunity ID'
               ]

value_variables = ['Community Served 1','Community Served 2', 
                   'Community Served 3', 'Community Served 4',
                   'Community Served 5'
                  ]

df = pd.melt(df, id_vars=id_variables, value_vars=value_variables, value_name='community', ignore_index=False)
df.shape
```

```python
# drop all the null values from the community column
# A null community means that this is not a real observation.
# Null communities should be dropped from the dataframe since
# they are just duplicate null observations. 

df = df[df['community'].notna()]
print(f'The shape of df is {df.shape}')
print(f"there are {df['community'].isnull().sum()} null values in the community column")
df.head()
```
